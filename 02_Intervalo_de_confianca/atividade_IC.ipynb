```python
def generate_samples(mu, size):
    samples = []
    for i in range(100): samples.append(poisson.rvs(mu=mu,size=size))
    return np.array(samples)
```

### **Exercício 2**

**Simule os intervalos de confiança para a média em cada uma das 100 amostras geradas no exercício anterior e entenda o significado do nível de confiança.**

O intervalo de confiança para a média (sabendo a distribuição da população), pelo Teorema Central do Limite, será dado por:

$
    \bar{x} - \dfrac{z_{\alpha/2}\sigma}{\sqrt{n}} \leq \mu \leq \bar{x} + \dfrac{z_{\alpha/2}\sigma}{\sqrt{n}}
$

Onde $\alpha$ é o nível de confiança. Faremos as simulações considerando $\alpha = 0.01$

```python
def interval_check(x, mu, var, n, alpha):
    L, U = norm.interval(1-alpha, loc=mu, scale=np.sqrt(var/n))
    leq = (x <= L) 
    ueq = (x >= U)
    out_of_limit = sum(leq)+sum(ueq)
    print(f'---------------------------------------------------------------------------------------')
    print(f'Amostra de tamanho {n}')
    print(f'Intervalo de confiança (alpha = {alpha}): [{L}, {U}]')
    print(f'Médias amostrais fora do intervalo de confiança: {out_of_limit}')
```

#### **a.** Para distribuição de Poisson com $\mu = 0.5$

```python
alpha = 0.05
l,u = norm.interval(1-alpha, loc=0, scale=1) 
```

Para essa primeira distribuição, calcularemos através da função *interval* do python e também manualmente, através da fórmula. Nas outras distribuições será utilizada apenas a função *interval*.

```python
L_20_a = 0.5 + l*np.sqrt(0.5/20)
U_20_a = 0.5 + u*np.sqrt(0.5/20)

print(L_20_a,U_20_a)
print(norm.interval(1-alpha, loc=0.5, scale=np.sqrt(0.5/20)))
```

```python
interval_check(a_20.mean(axis=1), 0.5, 0.5, 20, alpha)
interval_check(a_100.mean(axis=1), 0.5, 0.5, 100, alpha)
interval_check(a_500.mean(axis=1), 0.5, 0.5, 500, alpha)
interval_check(a_1000.mean(axis=1), 0.5, 0.5, 1000, alpha)
```

#### **b.** Para distribuição de Poisson com $\mu = 1.0$

```python
interval_check(b_20.mean(axis=1), 1.0, 1.0, 20, alpha)
interval_check(b_100.mean(axis=1), 1.0, 1.0, 100, alpha)
interval_check(b_500.mean(axis=1), 1.0, 1.0, 500, alpha)
interval_check(b_1000.mean(axis=1), 1.0, 1.0, 1000, alpha)
```

#### **c.** Para distribuição de Poisson com $\mu = 10.0$

```python
interval_check(c_20.mean(axis=1), 10.0, 10.0, 20, alpha)
interval_check(c_100.mean(axis=1), 10.0, 10.0, 100, alpha)
interval_check(c_500.mean(axis=1), 10.0, 10.0, 500, alpha)
interval_check(c_1000.mean(axis=1), 10.0, 10.0, 1000, alpha)
```

O significado do nível de confiança diz respeito à probabilidade da média de uma amostra estar dentro do intervalo adotado.
